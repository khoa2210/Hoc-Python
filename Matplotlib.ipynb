Line box 

```python
import matplotlib.pyplot as plt, numpy as np
plt.figure(figsize=(15,5)) # Giá trị dài và rộng biểu đồ 
plt.plot([1,2,3,4],[1,3,5,7]) # Giá trị X,Y
plt.show() #Vẽ hình 
```

Box plot

```python
x = np.random.randn(30) + np.arange(0, 30)*0.75 + 10
y = np.random.randn(10) + np.arange(0, 10)*1.0 +20
z = np.random.randn(10) + np.arange(0, 10)*0.5 + 5
t = np.random.randn(40) + np.arange(0, 40)*0.4 + 0.0
# Các giá trị ngẫu nhiên x,y,z,t 
plt.boxplot([x, y, z, t],
            labels = ['Mùa Xuân', 'Mùa Hạ', 'Mùa Thu','Mùa Đông'],
            showfliers = True) # các giấ trị labels

plt.title('''bieu do the hien su bien thien nhiet do''')# tiêu đề
plt.xlabel('Mùa') # Giá trị X của bảng
plt.ylabel('Chỉ số nhiệt độ ') #Giá trị Y của bảng
plt.show()
```

Bar plot

```python
color_names = ["Hung", "Hai", "Hung", "Khanh", "Long"] # bảng tên
colors = ["#3498db", "#fd79a8", "#f1c40f", "#273c75", "#e74c3c"]# bảng màu
num_favorite = [10,3,4,5,6] # giá trị tương ứng với bảng tên
xs = [i + 0.03 for i, _ in enumerate(color_names)] 
plt.bar(xs, num_favorite, color=colors) # vẽ chỉ số tên,giá trị và màu
plt.title("Hoc sinh lop 9") # tên bảng
plt.xticks([i+0.1 for i, _ in enumerate(color_names)], color_names)  # Tên các giá trị
plt.xlabel("Ten") # Giá trị X của bảng
plt.ylabel("Diem") # Giá trị Y của bảng
plt.show()
```

Scatter plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) 
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
# giá trị x và y của từng biến
plt.scatter(x, y)
plt.show()
```

Vẽ đồ thị một số hàm toán học 

```python
x = np.linspace(1, 100, 50)
 
fig = plt.figure(figsize = (14, 8))
 
# hàm y = x
y = x
plt.plot(x, y, 'b', label ='x')
 
# hàm bậc 2 
y2 = x*x-2*x+1
plt.plot(x, y2, 'r-.', label ='hàm mũ 2')
 
# hàm bậc 3
y3 = x*x*x-3*x*x+x
plt.plot(x, y3, 'g:', label ='Hàm mũ 3')
 
#Thêm feature
plt.legend()
plt.grid(True, linestyle =':')
plt.xlim([0, 50])
plt.ylim([0, 40])
 
plt.title('Hàm toán học')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
 
# Vẽ hình
plt.show()
```
